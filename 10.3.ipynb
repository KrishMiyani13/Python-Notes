```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,9))
x = np.random.randint(1000,size=(5))
my = ['Apple',"Banana","Mango","Cherry","Dates"]
my1 = ['brown','pink','yellow','orange','red']
my2 = my1[::-1]
myexplode = [0,.2,0,.05,.1]
plt.pie(x,startangle=90,labels=my ,colors=my2,explode=myexplode,shadow=True,radius=1,autopct='%1.4f%%')
plt.legend(loc='upper left',title='Fruits')
plt.show()
```

# Area Chart

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
x = np.random.randint(11,size=(5))
y = np.random.randint(11,size=(5))
plt.fill_between(x,y,color="brown",alpha=0.4)
plt.xlabel("Average Palse")
plt.ylabel("Calorie Burnage")
plt.title("Sports Watch Data")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
x = np.random.randint(1,11,size=(10))
y = np.random.randint(1,11,size=(10))
z = np.arange(10)
plt.plot(z,y,color="green")
plt.plot(z,x,color="Red")
plt.fill_between(z,y,x,where=(y>x),color='green',alpha = 0.25,label='Positive',interpolate=True)
plt.fill_between(z,y,x,where=(y<x),color='red',alpha = 0.25,label='Negative',interpolate=True)
plt.xlabel("Time")
plt.ylabel("Income and Expence")
plt.title("Area Fill between two line")
plt.legend()
plt.show()
```
