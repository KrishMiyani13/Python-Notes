# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0,6])
y = np.array([0,10])
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0,6])
y = np.array([0,10])
plt.plot(x,y,'--r',)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
x = np.random.randint(-10,11,size=(5))
y = np.random.randint(-10,11,size=(5))
plt.plot(x,y,'o-.r',ms=10,mec="b",mfc='g')
plt.plot(y,x,'o-.b',ms=10,mec="r",mfc='y')
plt.show()
```
