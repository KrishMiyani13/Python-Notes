```python
# Concatenate()
import numpy as np
arr1 = np.array([[[1,2,3],[2,3,4]]])
arr2 = np.array([[[4,5,6],[5,6,7]]])
arr3 = np.array([[[7,8,9],[8,9,10]]])
arr = np.concatenate((arr1,arr2,arr3))
print(arr)
```

```python
arr1 = np.array([[[1,2,3],[2,3,4]]])
arr2 = np.array([[[4,5,6],[5,4,7]]])
arr3 = np.array([[[7,8,9],[8,9,10]]])
arr = np.concatenate((arr1,arr2,arr3))
y = np.where(arr%2!=0)
print(y)
```
